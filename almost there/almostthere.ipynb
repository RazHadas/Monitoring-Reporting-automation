```python
def make_window(theme):
    sg.theme(theme)
    menu_def = [['&Application', ['E&xit']],
                ['&Help', ['&About']] ]
    right_click_menu_def = [[], ['Edit Me', 'Versions', 'Nothing','More Nothing','Exit']]
    graph_right_click_menu_def = [[], ['Erase','Draw Line', 'Draw',['Circle', 'Rectangle', 'Image'], 'Exit']]

    # Table Data
    data = [["John", 10], ["Jen", 5]]
    headings = ["Name", "Score"]

    input_layout =  [

                # [sg.Menu(menu_def, key='-MENU-')],
                [sg.Text('קובץ נותני שירותים פיננסים')], 
                [sg.Input(key='-FSP INPUT-')],
                [sg.Text('קובץ מדינות')], 
                [sg.Input(key='-RC INPUT-')],
                [sg.Checkbox('GMT', default=True, k='-GMT-')],
                [sg.Checkbox('yeshut', default=True, k='-yeshut-')],
                [sg.Checkbox('WIC', default=True, k='-WIC-')],
                [sg.Checkbox('Changemat', default=True, k='-CM-')],
                [sg.Checkbox('cox', default=True, k='-cox-')],
                [sg.Button('RUN Yeshut', size=(8, 4) ), sg.Button('read directory', size=(8, 4)),
                 sg.Button('read template directory', size=(8, 4)),
                 sg.Button('read fixed filies directory', size=(8, 4)),
                 sg.Button('RUN GMT', size=(8, 4) ),
                 sg.Button('RUN Changemat', size=(8, 4) ),
                 sg.Button('RUN RIA', size=(8, 4) ),
                 sg.Button('RUN STB', size=(8, 4) )]]

    
    theme_layout = [[sg.Text("See how elements look under different themes by choosing a different theme here!")],
                    [sg.Listbox(values = sg.theme_list(), 
                      size =(20, 12), 
                      key ='-THEME LISTBOX-',
                      enable_events = True)],
                      [sg.Button("Set Theme")]]
    
    layout = [ [sg.MenubarCustom(menu_def, key='-MENU-', font='Courier 15', tearoff=True)],
                [sg.Text('מערכת ניטור דיווחים בלתי רגילים', size=(38, 1), justification='center', font=("Helvetica", 16), relief=sg.RELIEF_RIDGE, k='-TEXT HEADING-', enable_events=True)]]
    layout +=[[sg.TabGroup([[  sg.Tab('ראשי', input_layout),
                               
                               
                               
                               sg.Tab('Theming', theme_layout),
                               ]], key='-TAB GROUP-', expand_x=True, expand_y=True),

               ]]
    layout[-1].append(sg.Sizegrip())
    window = sg.Window('מערכת ניטור דיווחים בלתי רגילים', layout, right_click_menu=right_click_menu_def, right_click_menu_tearoff=True, grab_anywhere=True, resizable=True, margins=(0,0), use_custom_titlebar=True, finalize=True, keep_on_top=True)
    window.set_min_size(window.size)
    return window

def main():
    window = make_window(sg.theme('bluemono'))
    
    # This is an Event Loop 
    while True:
        event, values = window.read(timeout=100)
        # keep an animation running so show things are happening
        
        if event not in (sg.TIMEOUT_EVENT, sg.WIN_CLOSED):
            print('============ Event = ', event, ' ==============')
            print('-------- Values Dictionary (key=value) --------')
            for key in values:
                print(key, ' = ',values[key])
        if event in (None, 'Exit'):
            print("[LOG] Clicked Exit!")
            break
        elif event == 'About':
            print("[LOG] Clicked About!")
            sg.popup('PySimpleGUI Demo All Elements',
                     'Right click anywhere to see right click menu',
                     'Visit each of the tabs to see available elements',
                     'Output of event and values can be see in Output tab',
                     'The event and values dictionary is printed after every event', keep_on_top=True)
            
        elif event == "read directory":
            
            directory = sg.popup_get_folder('Choose your folder', keep_on_top=True)
            text_file = open("directory.txt", "w")
 
            #write string to file
            text_file.write(directory)
 
            #close file
            text_file.close()
            sg.popup("You chose: " + str(directory), keep_on_top=True)
            
        elif event == "read fixed filies directory":
            fixed_directory = sg.popup_get_folder('Choose your folder', keep_on_top=True)
            text_file = open("perma directory.txt", "w")
 
            #write string to file
            text_file.write(fixed_directory)
 
            #close file
            text_file.close()
            sg.popup("You chose: " + str(fixed_directory), keep_on_top=True)            
            
        elif event == "read template directory":
            template_directory = sg.popup_get_folder('Choose your folder', keep_on_top=True)
            text_file = open("template directory.txt", "w")
 
            #write string to file
            text_file.write(template_directory)
 
            #close file
            text_file.close()
            sg.popup("You chose: " + str(template_directory), keep_on_top=True)            
            
                        
        elif event == 'RUN Yeshut': 
            sg.popup("מבצע ניטור", keep_on_top=True)         
            
            #running the check monitoring script
            %run yeshut.ipynb
            sg.popup("Done!", keep_on_top=True)
            
        elif event == 'RUN STB':
            sg.popup("מבצע ניטור", keep_on_top=True)         
            
            #running the check monitoring script
            %run STB.ipynb
            sg.popup("Done!", keep_on_top=True)
            
        elif event == 'RUN RIA':
            
            sg.popup("מבצע ניטור", keep_on_top=True)         
            
            #running the check monitoring script
            %run RIA.ipynb
            sg.popup("Done!", keep_on_top=True)           

        elif event == 'RUN GMT':
            
            sg.popup("מבצע ניטור", keep_on_top=True)         
            
            #running the check monitoring script
            %run GMT.ipynb
            sg.popup("Done!", keep_on_top=True)           

                        
            
        elif event == "Set Theme":
            print("[LOG] Clicked Set Theme!")
            theme_chosen = values['-THEME LISTBOX-'][0]
            print("[LOG] User Chose Theme: " + str(theme_chosen))
            window.close()
            window = make_window(theme_chosen)

        elif event == 'FSP input':
            sg
    window.close()
    exit(0)

if __name__ == '__main__':
    sg.theme('bluemono')
    sg.theme('dark red')
    sg.theme('dark green 7')
    # sg.theme('DefaultNoMoreNagging')
    main()
```
